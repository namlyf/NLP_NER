```python
import requests
from bs4 import BeautifulSoup
import csv
import time
import re
from underthesea import sent_tokenize, word_tokenize,pos_tag
import pandas as pd
import unicodedata
```

```python
def scrape_article_body_from_url(article_url):
    """
    Truy cập vào URL chi tiết của bài báo và trích xuất Tiêu đề và Nội dung chính.
    """
    HEADERS = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        response = requests.get(article_url, headers=HEADERS, timeout=15)
        response.raise_for_status() # Báo lỗi nếu mã trạng thái không phải 200
        soup = BeautifulSoup(response.content, 'html.parser')

        # --- 1. TRÍCH XUẤT TIÊU ĐỀ (TITLE) ---
        title_tag = soup.find('h1', class_='article__title')
        title = title_tag.get_text(strip=True) if title_tag else "Không tìm thấy Tiêu đề"

        # --- 2. TRÍCH XUẤT NỘI DUNG (BODY) ---
        # Tìm kiếm container chứa nội dung bài viết chính (dựa trên các class phổ biến)
        content_container = soup.find('div', class_='article__body') or \
                            soup.find('div', class_='content-detail') or \
                            soup.find('div', id='inner-article')
        if not content_container:
            return "KHÔNG TÌM THẤY CONTAINER NỘI DUNG (article__body, content-detail, inner-article)"
        else:
            # Tìm tất cả các đoạn văn bản (p) bên trong container
            paragraphs = content_container.find_all('p')
            if len(paragraphs) > 0:
                paragraphs.pop(0)
            full_text = '\n'.join([p.get_text(" ",strip=True) for p in paragraphs if p.get_text(strip=True) and not p.find_parent('figcaption')])
            return full_text

    except requests.exceptions.HTTPError as e:
        print(f"❌ Lỗi HTTP: Không thể truy cập URL. Mã lỗi: {e.response.status_code}")
        return None

    except requests.exceptions.RequestException as e:
        print(f"❌ Lỗi kết nối khi tải bài viết: {e}")
        return None

ARTICLE_URL = "https://vietnamnet.vn/khoi-nguyen-con-trai-nsnd-hong-van-bo-my-ve-viet-nam-gio-ra-sao-2462726.html"
global article_data 
article_data = scrape_article_body_from_url(ARTICLE_URL)

print(article_data)
```

```python
def clean_text(text):

    text = str(text)
    text = unicodedata.normalize("NFC", text)  # chuẩn hóa Unicode       
    text = re.sub(r"[\r\t]", " ", text)        # xóa ký tự lạ
    text = re.sub(r"\s+", " ", text).strip()   # gom khoảng trắng
    return text


all_sentences = []  

text = article_data
text = clean_text(text)
print(article_data)
# Tách câu
sentences = sent_tokenize(text)

for sent in sentences:
    # Tokenize
    tokens = word_tokenize(sent)
        
    # Mặc định label = "O" (chưa annotate)
    sent_data = [(tok, "O") for tok in tokens]
    all_sentences.append(sent_data)

print(all_sentences)
```
